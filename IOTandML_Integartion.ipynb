```python
results = model.predict(source='/content/Farm-Animals.jpg', save=False)

object_counts = {}
for result in results:
    boxes = result.boxes
    for box in boxes:
        class_id = box.cls.item()

        class_name = model.names[class_id]
        object_counts[class_name] = object_counts.get(class_name, 0) + 1

s=''

# Print the object counts
for class_name, count in object_counts.items():
    s+=f"{class_name}: {count}"+"\n"

print(s)
```

```python
results = model.predict(source='/content/Farm-Animals.jpg', save=True)

object_counts = {}

for result in results:
    boxes = result.boxes
    for box in boxes:
        class_id = box.cls.item()

        class_name = model.names[class_id]
        object_counts[class_name] = object_counts.get(class_name, 0) + 1

json_data = json.dumps(object_counts, indent=4)


print(json_data)
```
